```python
# creating our first function
# using the features from movement and column
def artist_features(movement, gender):
    return {'movement': movement, 'gender':gender}
```

```python
# zip combines our two columns
# we are making tuples
# our tuple is the movement and the gender as thing 1 and thing 2
# syntax for tuple (thing1, thing2) (movement, gender)
# (["Pop Art", male], price)
zipped_features_and_labels = zip(dataset.movement, list_of_genders, dataset.price)
# once created the list_of_movements above, this would overwrite dataset.movement to list_of_movements
```

```python
# if we only wanted to use the first word of the movement, we could write out this syntax
# but I want my dataset to be as accurate as possible and include multiple movements so I have decided to split these above
# !! will come back to this
# feature_sets = [ (setup_feature(movement.split(',')[0], gender, price) for movement, gender, price in zipped_features_and_labels ]
feature_sets = [ (artist_features(movement, gender), price) for movement, gender, price in zipped_features_and_labels ]
feature_sets
```

```python
# always want to shuffle your dataset before classifying
import random
random.shuffle(feature_sets)
```

```python
# this is one of the ways to work out 80% of our total 754 values/rows
# math.floor rounds our int down
import math
math.floor(754 * .8)
```

```python
# spliting our training set and testing set into a 80/20 ratio
# training is about 'learning'
# testing is to determine accuracy
training_set = feature_sets[:603]
testing_set = feature_sets[603:]
```

```python
classifier = nltk.NaiveBayesClassifier.train(training_set)
```
